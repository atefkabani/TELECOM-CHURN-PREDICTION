```python
import pandas as pd
import joblib
```

```python
model=joblib.load("../model/log_reg_pipeline.joblib")
```

```python
# Create example customer data for testing the model

example_customer1 = pd.DataFrame({
    "gender": ["Male"],
        "SeniorCitizen": ["No"],
        "Partner": ["No"],
        "Dependents": ["No"],
        "tenure": [16],
        "PhoneService": ["Yes"],
        "MultipleLines": ["No"],
        "InternetService": ["No internet service"],
        "OnlineSecurity": ["No internet service"],
        "OnlineBackup": ["No internet service"],
        "DeviceProtection": ["No internet service"],
        "TechSupport": ["No internet service"],
        "StreamingTV": ["No internet service"],
        "StreamingMovies": ["No internet service"],
        "Contract": ["Two year"],
        "PaperlessBilling": ["No"],
        "PaymentMethod": ["Credit card (automatic)"],
        "MonthlyCharges": [18.95],
        "TotalCharges": [326.8]
})

example_customer2 = pd.DataFrame({
    "gender": ["Male"],
    "SeniorCitizen": ["No"],
    "Partner": ["No"],
    "Dependents": ["No"],
    "tenure": [49],
    "PhoneService": ["Yes"],
    "MultipleLines": ["Yes"],
    "InternetService": ["Fiber optic"],
    "OnlineSecurity": ["No"],
    "OnlineBackup": ["Yes"],
    "DeviceProtection": ["Yes"],
    "TechSupport": ["No"],
    "StreamingTV": ["Yes"],
    "StreamingMovies": ["Yes"],
    "Contract": ["Month-to-month"],
    "PaperlessBilling": ["Yes"],
    "PaymentMethod": ["Bank transfer (automatic)"],
    "MonthlyCharges": [103.7],
    "TotalCharges": [50360.3]
})

prediction1 = model.predict(example_customer1)[0]
prediction2 = model.predict(example_customer2)[0]
 
if prediction1 == 1:
    print("The first customer is likely to churn.")
else:
    print("The first customer is likely to stay.")

if prediction2 == 1:
    print("The second customer is likely to churn.")
else:
    print("The second customer is likely to stay.")
```
